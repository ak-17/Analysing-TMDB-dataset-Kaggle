```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("tmdb_5000_movies.csv")
df = df.drop(df.columns[[1,2,5,7,9,10,15,14,16]], axis=1)
df.head()
```

```python
df.info()
```

```python
df['revenue'] = df['revenue']/1000000
df['budget'] = df['budget']/1000000
df['profit'] = df['revenue']-df['budget']
df['% profit'] = (df['profit']/df['budget'])*100
df.head()
```

```python
plt.scatter(df['budget'],df['% profit'])
plt.xlabel('budget')
plt.ylabel('% profit')
plt.ylim([-100,0])
##plt.xlim([0,1000])
plt.ion()
plt.show()
```

```python
plt.plot(df['profit'],df['budget'])
plt.xlabel('budget')
plt.ylabel('profit')
##plt.xlim([0,1000])
plt.ion()
plt.show()
```
